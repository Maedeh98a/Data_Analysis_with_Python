```python
# export the plot into a file

import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 1, 1])

no_of_students = np.array([35, 30, 25, 45])
courses = ["Python", "java", "Data Science", "Big-Data"]
ax.pie(no_of_students, labels = courses)
ax.set_title("Pi Chart")

fig.savefig("D:\Downlods\myplot.jpeg", dpi=2000) #saving the plot in myplot.jpeg file
```

```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 1, 1])

no_of_students = np.array([35, 30, 25, 45])
courses = ["Python", "java", "Data Science", "Big-Data"]
ax.pie(no_of_students, labels = courses)
ax.set_title("Pi Chart")

mypdf = PdfPages('"D:\Downlods\myplotpdf.pdf')
mypdf.savefig(fig)
mypdf.close()
```
