```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 2, 3, 4]

plt.plot(x, y)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.5, 0.5]) # [left, bottom, width, height]
x = [1, 2, 3, 4]
y = [1, 2, 3, 4]
ax.plot(x, y)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2) # it means we want 1 rows and 2 columns
fig.set_facecolor('lightgray') # setting the display color for figure 
x = [1, 2, 3, 4, 5, 6]
y = [1, 2, 3, 4, 5, 6]
x_squre = [x*x for x in x]

axes[0].plot(x, y)
axes[1].plot(x, x_squre)

plt.show()
```

```python
# create a 2 by 2 subplot, meaning 4 subplots
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_facecolor('lightgray') 
x = [1, 2, 3, 4, 5, 6]
y = [1, 2, 3, 4, 5, 6]

x_squre = [x*x for x in x]
y_squre = [y*y for y in y]

axes[0][0].plot(x, y)
axes[0][0].set_title("I") # setting the title of subplot

axes[0][1].plot(x, x_squre)
axes[0][1].set_title("II")

axes[1][0].plot(y, y_squre)
axes[1][0].set_title("III")

axes[1][1].plot(x_squre, y_squre)
axes[1][1].set_title("IV")

plt.show()
```
