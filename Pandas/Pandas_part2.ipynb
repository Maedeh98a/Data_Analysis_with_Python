```python
#filling the missing values
import numpy as np
import pandas as pd
emp_dict = {
    'Name':['Porter', 'Abraham', 'Ali', 'Hasan', 'Sina', np.nan, 'Sepehr'],
    'Department':['Advertising', np.nan, 'Accounting', '', 'Accounting', np.nan, 'Bussiness'],
    'Salary_in_dollar':[7921, 7732, 7811, np.nan, 7443, np.nan, 0],
    'city':['Haarlem', 'qom', np.nan, 'Shiraz', 'Tehran', 'Tehran', "NA"]
}
df = pd.DataFrame(emp_dict)
df
```

```python
df_fillna = df.fillna("MissingValue")
print(df_fillna)
```

```python
df_fillna_w_ffil = df.fillna(method='ffill')
df_fillna_w_ffil
```

```python
sal_fillna_w_ffill = df[['Salary_in_dollar']].fillna(method='ffill')
sal_fillna_w_ffill
```

```python
df
```

```python
salary_w_bfill = df[['Salary_in_dollar']].fillna(method='bfill')
salary_w_bfill
```

```python
# fill the NAN with mean
df_fillna_w_mean = df[['Salary_in_dollar']].fillna(df['Salary_in_dollar'].mean())
df_fillna_w_mean
```

```python
df_fillna_w_median = df[['Salary_in_dollar']].fillna(df['Salary_in_dollar'].median())
df_fillna_w_median
```

```python
# why using mode in this way?
df_fillna_w_mode = df[['Salary_in_dollar']].fillna(df['Salary_in_dollar'].mode()[0])
df_fillna_w_mode
```

```python
#why use 0 as index?
df_fillna_w_mode = df[['city']].fillna(df['city'].mode()[0])
df_fillna_w_mode
```

```python
df_replace = df.replace({'NA': 'Not applicable', 0:5000,"":'Blank', np.nan:'Not a Number'})
df_replace
```

```python
salary_replaced = df[['Salary_in_dollar']].replace({0:5000, np.nan:5000})
salary_replaced
```
