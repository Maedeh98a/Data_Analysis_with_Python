```python
import pandas as pd
import numpy as np
#sample data frame
emp_dict = {'emp_id':[101, 105, 108, 100],
           'city':['ahvaz', ' yazd', 'qom', 'aras'],
           'sal':[5000, 8989, 5634, 9634]}
df = pd.DataFrame(emp_dict, index=[1, 4, 2, 3])

#sorting the data using sort index
df_sorted_on_index = df.sort_index()
df
```

```python
df_sorted_on_index
```

```python
# sort by columns value
df_sorted_on_values = df.sort_values(['emp_id'])
print("Using sort_values() with default")
df_sorted_on_values
```

```python
df_sorted_on_sal = df.sort_values(['sal'], ascending=False)
df_sorted_on_sal
```

```python
df['sal_rank_asc'] = df['sal'].rank()
df
```

```python
df['sal_rank_desc'] = df['sal'].rank(ascending=False)
df
```

```python
# Adding Row
# We have the previous dataframe so we use it again
row = {'emp_id': 999, 'city': 'Haarlem', 'sal': 6789}
print(type(row))
df = df.append(row, ignore_index=True)
df
```

```python
# adding a new raw, when new row type is pandas series
row = pd.Series(data = {'emp_id': 888, 'city':'rom', 'sal': 6789}, name=6)
print(type(row))
df = df.append(row, ignore_index=False)
df
```

```python
# adding a column with constant singel value for each row in dataframe
df['constant_value_col'] = 'New_value'
df
```

```python
# Adding column into dataframe
df['new_col'] = [1, 2, 4, 3, 6, 5]
df
```

```python
# writing our python function by apply function

def sal_with_10_percent_hike(old_sal):
    return int(old_sal) + int(old_sal) * 0.1

#adding new column updated_salary by apply the function sal_with_10_percent_hike
df['updated_salary'] = df['sal'].apply(sal_with_10_percent_hike)
df
```

```python
df['updated_salary'] = df['sal'].apply(lambda sal: int(sal) + int(sal) * 0.1)
df
```
