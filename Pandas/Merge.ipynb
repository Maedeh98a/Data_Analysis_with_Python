```python
import pandas as pd
import numpy as np
emp_dict = {'emp_id':[101, 105, 108, 100],
           'city':['ahvaz', ' yazd', 'qom', 'aras']}
dept_dict = {'emp_id':[101, 105, 108, 111],
            'dept':['Customer', 'legal', 'public', 'advertising']}

emp_df = pd.DataFrame(emp_dict)
dept_df = pd.DataFrame(dept_dict)



# how to display both of them????????????
emp_df
dept_df
```

```python
merged_df_inner = pd.merge(emp_df, dept_df, on='emp_id', how='inner')
merged_df_inner
```

```python
merged_df_left = pd.merge(emp_df, dept_df, on='emp_id', how='left')
merged_df_left
```

```python
merged_df_right = pd.merge(emp_df, dept_df, on='emp_id', how='right')
merged_df_right
```

```python
merged_df_outer = pd.merge(emp_df, dept_df, on='emp_id', how='outer')
merged_df_outer
```
